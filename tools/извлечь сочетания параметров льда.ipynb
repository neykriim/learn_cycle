```python
import os
import glob
import pandas as pd
import geopandas as gpd
from tqdm import tqdm
import numpy as np

ice_parameters = ["CT", "CA", "CB", "CC", "SA", "SB", "SC", "CN", "CD", "FA", "FB", "FC", "CF"]

ice_types = pd.read_csv('ice_types.csv', index_col=0, dtype=str)

images = tqdm(glob.glob("../прыг/*.gpkg"))

for image in images:
    if not os.path.isfile(os.path.basename(image)):
        data = gpd.read_file(image)
        images.set_description(image)
        if set(ice_parameters).issubset(data.columns):
            data['ice'] = ""
            data.loc[data[ice_parameters].isna().all(axis=1), ice_parameters] = '-9'
            unique = data[ice_parameters].drop_duplicates()#.dropna(axis=0, how="all")
            for _, ice_type in unique.iterrows():
                mask1 = (ice_types[ice_parameters] == ice_type).all(axis=1)
                mask2 = (data[ice_parameters] == ice_type).all(axis=1)
                if mask1.any():
                    #print("Сочетание найдено!")
                    # Получаем индексы строк, где встречается это сочетание
                    ice_index = ice_types[mask1].index.tolist()[0]
                    #print("Индексы строк:", ice_index)
                else:
                    #print("Сочетание не найдено.")
                    ice_index = len(ice_types)
                    ice_types.loc[ice_index] = ice_type

                data.loc[mask2, 'ice'] = ice_index

            if not '' in data['ice'].values:
                data = data.drop(columns=ice_parameters)
            else:
                print(f"Не все сочетания были определены в слое {image}")
            data.to_file(os.path.basename(image))

            ice_types.to_csv('ice_types.csv')
        else:
            pass
            #print(f"{image} не имеет параметров льда")
```
